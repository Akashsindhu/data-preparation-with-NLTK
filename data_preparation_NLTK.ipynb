```
filename = 'metamorphosis.txt'
file = open(filename, 'rt')
text = file.read()
file.close()
```

## Split by Whitespace

```
words = text.split()
words[:100]
```

## using regex model (re) to split the document into words.

```
import re
words = re.split(r'\W+', text)
words
```

```
import string
print(string.punctuation)
```

```
re_punc = re.compile('[%s]' % re.escape(string.punctuation))
# remove punctuation from each word 
stripped = [re_punc.sub('', w) for w in words]
```

## Tokenization and cleaning with NLTK

## NLTK provides 


1.   split into sentences 
2.   split into words
3.   filter out punctuations 
4.   filter out stop words 
5.   stem words 





```
from nltk import sent_tokenize
sentences = sent_tokenize(text)
```

```
sentences
```

```
from nltk import word_tokenize
tokens = word_tokenize(text)
tokens
```

```
words = [word for word in tokens if word.isalpha()]
words
```

```
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
(stopwords)
```

```
words = [w for w in words if not w in stopwords]   # this is the best line 
# i have ever seen, it means if w in words is not in the stopwords, then print it 
len(words)
```

```
len(stopwords)
```

### famous long standing method is Porter Stemming algorithm. 

```
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
words = [porter.stem(word) for word in words]
words 
```

```
len(words)
```

### Using sklearn to prepare text data 

### Word Counts with CountVectorizer


```
from sklearn.feature_extraction.text import CountVectorizer

text = ['This quick brown fox jumped over the lazy dog.']

token = CountVectorizer()
token.fit(text)
print(token.vocabulary_)

encode = token.transform(text)
print(encode.shape)
print(type(encode))
print(encode.toarray())
```

```
text2 = ['This dog is not lazy.']
predict = token.transform(text2)
print(predict.toarray())
```

### Word frequency with tfidfvectorizer

Term Frequency: This summarizes how often a given word appears within a document.

Inverse Document Frequency: This downscales words that appear a lot across documents.


```
from sklearn.feature_extraction.text import TfidfVectorizer
text = ['The quick brown fox jumped over the lazy dog.',
        'The dog',
        'The fox']
vector = TfidfVectorizer()
vector.fit(text)

print(vector.vocabulary_)
print(vector.idf_)

encode = vector.transform([text[0]])
print(encode.shape)
print(encode.toarray())
```

### Hashing with hashingVectorizer

```
from sklearn.feature_extraction.text import HashingVectorizer
text = ['The quick brown fox jumped over the lazy dog.']
vector = HashingVectorizer(n_features = 20)
encode = vector.transform(text)

# print(vector.vocabulary_)
print(encode.toarray())
```

### Prepare text with keras 

Split words with text_to_word_sequence. It do 3 things by default:-


1.   Split words by space
2.   Filter out punctuations
3.   convert text to lowercase



```
from keras.preprocessing.text import text_to_word_sequence

text = ['The quick brown fox jumped over the lazy dog.']

result = text_to_word_sequence(text[0])
result
```

### encoding with one_hot

```
vocab_size = len(result)
print(vocab_size)

from keras.preprocessing.text import one_hot

result = one_hot(text[0], round(vocab_size*1.3))
print(result)
```

### hash encoding with hashing_trick

```
from keras.preprocessing.text import hashing_trick
result = hashing_trick(text[0], round(vocab_size*1.3), hash_function = 'md5')
print(result)
```

### Tokenizer API

```
from keras.preprocessing.text import Tokenizer

docs = ['Well Done',
        'Good Work',
        'nice work',
        'Excellent!']

#  create a tokenizer 
t = Tokenizer()
t.fit_on_texts(docs)

# summarize what was learned 
print(t.word_counts)  # dictionary of words and their count 
print(t.document_count) # dictonary of words and how many documents each appeared in 
print(t.word_index) # dictionary of words and their uniquely assigned integers 
print(t.word_docs) # integer count of total no. of documents used in the fit tokenizer 

# integer encoded documents 
encoded_docs = t.texts_to_matrix(docs, mode = 'count')

print(encoded_docs)
```
